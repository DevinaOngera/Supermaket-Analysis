```python
#what time of the day do we have a higher sale volume
#Time of day when there are most sales in terms of quantity

Quantity = mydata["Quantity"]

Quantity

Time = mydata["Time"]

Quantitylist = list(Quantity)

Quantitylist

Timelist = list(Time)

Timelist

Newtimelist = []
#replacing : with . in the list so that we can be able to remove values after the decimal
for x in Timelist:
    mytime = x.replace(":", ".")
    #print(mytime)
    #float is number that carries decimal points. We use float to convert numbers with decimal points to whole numbers
    y = float(mytime)
    finaltime = math.floor(y)
    print(finaltime)
    Newtimelist.append(finaltime)

Newtimelist

```

```python


Newtimelist
```

```python
frequencylist = list(zip(Quantitylist, Newtimelist))

frequencylist
```
